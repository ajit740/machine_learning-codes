# Spotify Streaming History Exploratory data analysis(EDA) project

## Overview

This project focuses on performing Exploratory Data Analysis (EDA) on my Spotify streaming history data.
The analysis aims to uncover patterns, trends, and insights into my listening habits over different years.

## Project Structure

The project is organized into the following sections:

1. **Data Loading:**

   - The Spotify streaming history data is loaded from separate datasets representing different years.

2. **Data Cleaning:**

   - Missing values are handled appropriately.
   - Timestamps are converted to datetime objects.
   - Anomalies in the data are explored and cleaned.

3. **Data Exploration:**

   - Distribution analysis of streaming durations, platforms, and countries.
   - Identification of top tracks, artists, and albums.
   - Analysis of streaming habits over time.

4. **Visualization:**

   - Creation of visualizations using Matplotlib and Seaborn to represent findings.

5. **Summary and Insights:**

   - Key insights and observations drawn from the analysis.

# Data exploration

```python
#streaming duration
plt.figure(figsize=(12,6))
sns.histplot(df["ms_played"],bins=50,kde=True)
plt.title("striming duration")
plt.xlabel("milisecond played")
plt.ylabel("freequency")
plt.show()
```

```python
streaming_duration=df["master_metadata_album_artist_name"].value_counts().head(10)
print(streaming_duration)
```

```python
#no of platform
plt.figure(figsize=(10,6))
sns.countplot(x='platform',data=df)
plt.title("streaming platforms")
plt.xlabel("platform")
plt.ylabel("count")
plt.xticks(rotation=90)
plt.show()
```

```python
#no of countries
plt.figure(figsize=(10,6))
sns.countplot(x='conn_country',data=df)
plt.title("streaming platforms")
plt.xlabel("Country")
plt.ylabel("count")
plt.xticks(rotation=90)
plt.show()
```

```python
#top 10 tracks
top_tracks=df["master_metadata_track_name"].value_counts().head(5)
print(top_tracks)
```

```python
#top 10 artist
top_artist=df["master_metadata_album_artist_name"].value_counts().head(5)
print(top_artist)
```

```python
#top 10 albums
top_albums=df["master_metadata_album_album_name"].value_counts().head(5)
print(top_albums)
```

```python
#top albums
plt.figure(figsize=(10,5))
top_albums.plot(kind="bar")
plt.title("top 10 most stramed albums ")
plt.xlabel("songs")
plt.ylabel("count")
plt.show()
```

```python
#top 10 artist
plt.figure(figsize=(10,5))
top_artist.plot(kind="bar")
plt.title("top 10 artist ")
plt.xlabel("songs")
plt.ylabel("count")
plt.show()
```
