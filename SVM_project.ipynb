```python
# lets conduct some preliminiary data preparation steps
# rescaling the variable
# splitting it to train and test
# try to understand why scaling is imp 
# try to print the summary of the data.
```

```python
x=df.drop('spam',axis=1)
```

```python
y=df.spam
```

```python
y
```

```python
from sklearn.preprocessing import scale
```

```python
X=scale(x)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
print(y_train.mean())
```

```python
print(y_test.mean())
```

```python
# lets built the svm model
from sklearn.svm import SVC
```

```python
help(SVC)
```

```python
# instance an object of class SVC
# note that we are using c = 1

model=SVC(C=1)
```

```python
# fit the model
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)

```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_true = y_test,y_pred = y_pred)
```

```python
# now print the accuracy of metrix
print("accuracy", metrics.accuracy_score(y_test,y_pred))
```

```python
print("precision",metrics.precision_score(y_test,y_pred))
```

```python
print("recall",metrics.recall_score(y_test,y_pred))
```
